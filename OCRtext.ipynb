```python
from pdf2image import convert_from_path
import easyocr
import numpy as np
import PIL
from PIL import ImageDraw
```

```python
reader = easyocr.Reader(['en'])
```

```python
images = convert_from_path('/content/drive/MyDrive/ddp/R550-QSG-800-72433-001-RevB_230216_161849.pdf')
```

```python
from IPython.display import display, Image
display(images[0])
```

```python
bounds = []
for img in images:
  bound = reader.readtext(np.array(img),min_size=0,slope_ths=0.2,ycenter_ths=0.7, height_ths=0.6, width_ths=0.8,decoder='beamsearch',beamWidth=10)
  bounds.append(bound)
bounds
```

```python
bounds_sorted = sorted(bounds, key=lambda x: x[0][0])
```

```python
bounds = reader.readtext(np.array(images[0]), min_size=0, slope_ths = 0.2, ycenter_ths=0.7, height_ths=0.6, width_ths=0.8, decoder='beamsearch', beamWidth=10)
bounds
```

```python
def draw_boxes(image, bounds, color = 'yellow', width=2):
  draw  = ImageDraw.Draw(image)
  for bound in bounds:
    p0,p1,p2,p3 = bound[0]
    draw.line([*p0,*p1,*p2,*p3,*p0], fill=color,width=width)
  return image

draw_boxes(images[0], bounds)
```
